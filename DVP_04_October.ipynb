## Numerical Numerical

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')                   # Load the 'tips' dataset

df
```

### bargraph

```python
plt.bar(df['smoker'],df['total_bill'])

plt.show()
```

## barplot

```python
sns.barplot(data=df,x='sex',y='tip',errorbar=('ci', True))

plt.show()
```

```python
sns.barplot(data=df,x='sex',y='tip',hue='time',errorbar=('ci', False))

plt.show()
```

## boxplot

```python
sns.boxplot(data=df,x='sex',y='tip')
plt.yticks(range(1,11))
plt.show()
```

```python
sns.boxplot(data=df,x='sex',y='tip',hue='time',palette='Oranges')
plt.yticks(range(1,11))
plt.show()
```

```python
sns.boxplot(data=df,x='sex',y='tip',hue='smoker',palette='Purples')
plt.yticks(range(1,11))
plt.grid(color='r',ls='--')
plt.show()
```

```python
df
```

```python
sns.boxplot(data=df,x='sex',y='tip',hue='day',palette='Purples')
plt.yticks(range(1,11))
plt.grid(color='r',ls='--')
plt.show()
```

## violinplot

```python
sns.violinplot(data=df,x='day',y='total_bill')

plt.show()
```

```python
sns.violinplot(data=df,x='day',y='total_bill',palette='Purples')
plt.show()
```

```python
sns.violinplot(data=df,x='day',y='total_bill',hue='time')
plt.show()
```

```python
sns.violinplot(data=df,x='day',y='total_bill',hue='sex')
plt.show()
```

## stripplot

```python
sns.stripplot(data=df,x='day',y='total_bill')
# plt.axhline(20,c='r')
# plt.axvline(2,c='r')

plt.show()
```

```python
sns.stripplot(data=df,x='tip',y='sex')
# plt.axhline(20,c='r')
# plt.axvline(2,c='r')

plt.show()
```

```python
sns.stripplot(data=df,x='day',y='total_bill')
plt.axhline(20,c='r')
plt.axvline(2,c='r')

plt.show()
```

## swarmplot

```python
sns.swarmplot(data=df,x='day',y='total_bill')
plt.axhline(20,c='k')

plt.show()
```

```python
sns.swarmplot(data=df,x='day',y='total_bill',hue='day')
plt.axhline(20,c='k')
plt.show()
```

#### crosstab()

```python
import pandas as pd
```

```python
pd.crosstab(df['sex'],df['day'])
```

```python
pd.crosstab(df['sex'],df['day']).plot(kind='bar')

plt.show()
```

```python
pd.crosstab(df['sex'],df['day']).plot(kind='bar')
plt.text(0,60,59)
plt.text(-0.1,12,10)

plt.text(-0.2,35,30)

plt.text(0.15,58,58)

plt.text(1,30,28)
plt.text(0.90,10,10)
plt.text(0.80,32,32)
plt.text(1.15,20,18)
plt.show()
```

```python
pd.crosstab(df['sex'],df['day']).plot(kind='bar',stacked=True)

plt.show()
```

### Stacked Bar Graph

```python
pd.crosstab(df['sex'],df['day'])
```

```python
pd.crosstab(df['sex'],df['day']).plot(kind='bar',stacked=True)

plt.show()
```

```python
pd.crosstab(df['sex'],df['day']).plot(kind='bar',stacked=True)

plt.text(0,20,30,color='white')
plt.text(0,33,10,color='white')
plt.text(0,70,59,color='white')
plt.text(0,120,58,color='white')

plt.text(1,20,32,color='white')
plt.text(1,33,9,color='white')
plt.show()
```
