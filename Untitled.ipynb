```python
# Creating a dictionary comprehension to generate a dictionary of squares
numbers = [1, 2, 3, 4, 5]
squared_dict = {num: num ** 2 for num in numbers}

print(squared_dict)  # Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = [30, 40, 20, 10]


colors = ['red', 'blue', 'green', 'orange']


fig, ax = plt.subplots(figsize=(6, 6))
_, _, autopcts = ax.pie(data, labels=['Label 1', 'Label 2', 'Label 3', 'Label 4'], colors=colors, autopct='%1.1f%%')


for autopct in autopcts:
    autopct.set_color('white')  # Change 'white' to your desired label text color

# Add a title to the chart
ax.set_title('Pie Chart with Custom Percentage Label Color')

# Display the chart
plt.show()
```
