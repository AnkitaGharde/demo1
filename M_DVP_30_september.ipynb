```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')                   # Load the 'tips' dataset

df
```

## Scatterplot

#### example:

```python
plt.scatter(df['total_bill'],df['tip'])

plt.xlabel('Total bill($)')
plt.ylabel('tip($)')
plt.title('Relationship between total_bill & tip')

plt.show()


# Create a scatter plot to visualize the relationship between 'total_bill' and 'tip' columns
# Scatter plots show how two numerical variables are related by plotting points on a two-dimensional plane
```

```python
#add edgecolor

plt.scatter(df['total_bill'],df['tip'],edgecolor='r')


plt.xlabel('Total bill($)')
plt.ylabel('tip($)')
plt.title('Relationship between total_bill & tip')

plt.show()

# Scatter plots show how two numerical variables are related by plotting points on a two-dimensional plane
# 'edgecolor' sets the color of the marker edges to red ('r')
```

```python
plt.scatter(df['total_bill'],df['tip'],edgecolor='r',color='g')

plt.xlabel('Total bill($)')
plt.ylabel('tip($)')
plt.title('Relationship between total_bill & tip')
plt.axhline(1,color='black',ls='--') 
plt.axhline(4,color='black',ls='--') 
plt.axvline(10,color='black',ls='--') 
plt.axvline(25,color="g",ls='-')
plt.show()
```

```python
plt.scatter(df['total_bill'],df['tip'],edgecolor='r',color='g')
plt.xlabel('Total bill($)')
plt.ylabel('tip($)')
plt.title('Relationship between total_bill & tip')

plt.axhline(2,color='black',ls='--')  # Add horizontal dashed lines at y=2 and y=4 with black color
plt.axhline(4,color='k',ls='--')
plt.axvline(10,c='r',ls='--')# Add vertical dashed lines at x=10 and x=20 with red color
plt.axvline(20,c='r',ls='--')
plt.yticks(range(1,11))
plt.xticks(range(5,51,5))

plt.show()# Display the scatter plot with added lines
```

### using sns

```python
# #day
plt.figure(figsize=(6,4))

sns.scatterplot(data=df,x='total_bill',y='tip') 
plt.grid()
plt.show()
```

```python
#smoker

plt.figure(figsize=(6,4))

sns.scatterplot(data=df,x='total_bill',y='tip',hue="smoker") 
plt.grid()
plt.show()
```

```python
#sex

plt.figure(figsize=(6,4))

sns.scatterplot(data=df,x='total_bill',y='tip',hue="sex") 

plt.show()
```

```python
# #time
plt.figure(figsize=(6,4))
sns.scatterplot(data=df,x='total_bill',y='tip',hue='time')
plt.show()
```

### lmplot(Linear model plot)

```python
sns.lmplot(data=df,x='total_bill',y='tip')

plt.show()
```

```python
sns.lmplot(data=df,x='total_bill',y='tip',hue='smoker')

plt.show()
```
