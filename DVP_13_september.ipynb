```python
import pandas as pd
```

```python
c=pd.crosstab(df['sex'],df['day']).plot(kind='bar')
# plt.text(0,60,59)
# plt.text(-0.1,12,10)
# plt.text(-0.2,35,30)
# plt.text(0.15,58,58)

# plt.text(1,30,59)

for bars in c.containers:
        c.bar_label(bars)
plt.show()
```

```python
pd.crosstab(df['sex'],df['day'])
```
