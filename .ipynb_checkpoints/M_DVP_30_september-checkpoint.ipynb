### Side by side bar graph

#### crosstab()

```python
df
```

```python
import pandas as pd
```

```python
df['sex'].value_counts()
```

```python
df['day'].value_counts()
```

```python
pd.crosstab(df['sex'],df['day'])
```

```python
pd.crosstab(df['sex'],df['day']).plot(kind='bar')
plt.text(0,60,59)
plt.text(-0.1,12,10)
plt.text(-0.2,35,30)
plt.text(0.15,58,58)

plt.text(1,30,59)
plt.show()
```

### Stacked Bar Graph

```python
pd.crosstab(df['sex'],df['day'])
```

```python
pd.crosstab(df['sex'],df['day']).plot(kind='bar',stacked=True)
plt.text(0,18,30,color='white')
plt.text(0,33,10,color='white')
plt.text(0,70,59,color='white')
plt.text(0,120,58,color='white')
```
