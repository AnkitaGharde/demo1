## Numerical Numerical

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')                   # Load the 'tips' dataset

df
```

### lmplot(Linear model plot)

```python
sns.lmplot(data=df,x='total_bill',y='tip')

plt.show()
```

```python
sns.lmplot(data=df,x='total_bill',y='tip',hue='smoker')

plt.show()
```

## regplot

```python
# example1

sns.regplot(data=df,y='size',x='total_bill')
```

```python
# example2

sns.regplot(data=df,y='size',x='total_bill',hue='smoker')   # in regplot hue not support
plt.show()
```

## pairplot

## jointplot

```python
# example1

# plt.figure(figsize=(4,3))
sns.jointplot(data=df,x='size',y='tip')
plt.show()
```

```python
# example2

sns.jointplot(data=df,x='size',y='tip',hue='time')

plt.show()
```

## line

```python
# Sample data
x_values = [5,6,7,8,9]
y_values = [10, 20, 15, 25, 30]

plt.plot(x_values, y_values, marker='*', linestyle='--')
# plt.xticks(range(1,11,2))
# plt.yticks(range(1,31,3))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Line Plot')
plt.show()                                                           
```

```python
import pandas as pd

data = pd.DataFrame({'X': [1, 2, 3, 4, 5], 'Y': [10, 7, 12, 8, 15]})


sns.lineplot(x='X', y='Y', data=data, marker='*',mec='b',color='r')

plt.show()
```
